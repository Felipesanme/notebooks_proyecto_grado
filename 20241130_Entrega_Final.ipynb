## 3. Antecedentes
    
    Para este trabajo se tuvieron en cuenta estudios anteriores realcionados con el estudio de Pruebas Saber 11 y Saber Pro que utilizando tecnicas de machine Learning como el K-means, arboles de desicion, redes neuronales, han logrado evidenciar la influencia de la estadistica aplicada en el campo educativo. 
    
    Carrascal y Giraldo (2019) realizaron un estudio titulado Minería de Datos Educativos: Análisis del Desempeño de Estudiantes de Ingeniería en las Pruebas Saber-Pro. El objetivo de este trabajo fue contribuir al análisis de las pruebas Saber-Pro en Colombia, examinando variables académicas, económicas, sociodemográficas y los resultados de todos los módulos de dichas pruebas correspondientes al año 2016, específicamente de estudiantes de ingeniería en Antioquia. Para ello, emplearon tres modelos analíticos de clustering, identificaron los factores más influyentes en el desempeño de cada módulo de las pruebas de manera individual, y predijeron los resultados en función de las variables seleccionadas.

    Entre los hallazgos más destacados, se identificó que el género fue un factor determinante en los módulos de razonamiento cuantitativo y comunicación escrita, aunque tuvo menor incidencia en el desempeño de los demás módulos. Además, el número de libros en casa resultó ser un factor clave para la comunicación escrita, mientras que la condición económica influyó significativamente en el módulo de inglés. Por otro lado, se encontró que ser cabeza de familia representaba una barrera importante para obtener buenos resultados en las pruebas. Finalmente, el estudio también evidenció que los estudiantes que cursan sus estudios de manera presencial tienden a obtener mejores resultados, un aspecto que coincide con los hallazgos de investigaciones previas.

    Timarán, Caicedo y Hidalgo (2019) llevaron a cabo un estudio titulado Árboles de Decisión para Predecir Factores Asociados al Desempeño Académico de Estudiantes de Bachillerato en las Pruebas Saber 11°. El objetivo fue identificar los factores que influyen en el desempeño académico de los estudiantes colombianos de grado undécimo que presentaron las pruebas Saber 11° en los años 2015 y 2016.

    Para descubrir patrones relacionados con el desempeño académico, los investigadores desarrollaron un modelo de clasificación basado en árboles de decisión, empleando el algoritmo J48 de la herramienta WEKA y siguiendo la metodología CRISP-DM. El modelo, construido con la base de datos T1061680A17, logró clasificar correctamente 711,116 instancias, alcanzando una precisión del 67%, mientras que 350,564 instancias fueron clasificadas incorrectamente, representando un 33% de error.

    Los autores concluyeron que los árboles de decisión son herramientas efectivas para generar modelos consistentes con la realidad observada, respaldados teóricamente y basados únicamente en los datos almacenados en las bases de datos del ICFES. Este enfoque proporciona valiosa información para analizar y predecir el desempeño académico en las pruebas Saber 11°.

    Rojas (2021) llevó a cabo un estudio titulado Comparación de Herramientas Libres de Minería de Datos Educativa que Aplican Técnicas Descriptivas. Caso de Estudio: Caracterización Socioeconómica de los Estudiantes en las Pruebas Saber 11 de 2018 y 2019. El objetivo fue comparar herramientas de minería de datos educativa de código abierto que aplican técnicas descriptivas, con el propósito de caracterizar la situación socioeconómica de los estudiantes que presentaron las pruebas Saber 11 en los años mencionados.

    Para ello, implementó una metodología comparativa de las herramientas KNIME, Orange y WEKA, siguiendo el proceso de minería de datos educativa guiado por la metodología CRISP-DM. Entre los hallazgos, se identificaron como factores clave para un mejor desempeño académico el acceso a bienes y servicios, particularmente un computador y conexión a internet, así como el nivel educativo de los padres. Los estudiantes cuyos padres alcanzaron niveles educativos superiores al bachillerato obtuvieron mejores puntajes. Por otro lado, aquellos con bajo rendimiento presentaron características como mayor edad, convivir con más de siete personas en el hogar, poseer motocicleta, trabajar más de 10 horas a la semana y recibir remuneración en efectivo.

    La implementación de los algoritmos K-Means y Jerárquico en las tres herramientas mostró resultados consistentes, agrupando los clústeres según los puntajes obtenidos en las diferentes áreas de las pruebas (bajos, medios y altos) y las variables relacionadas con dichos puntajes. Este enfoque destacó la capacidad de las herramientas para analizar y categorizar factores socioeconómicos relevantes en el desempeño académico.

## 5. Descripción detallada de los datos
  
    Las variables consideradas para el estudio incluyen aspectos relacionados con el colegio, el estudiante y su entorno familiar. Entre las variables del colegio se encuentran: el área de ubicación (COLE_AREA_UBICACION), el tipo de calendario académico (COLE_CALENDARIO), el nombre de la sede (COLE_NOMBRE_SEDE), el carácter del colegio (público o privado, COLE_CARACTER), la jornada académica (mañana, tarde o completa, COLE_JORNADA), el género predominante en el colegio (mixto, masculino o femenino, COLE_GENERO) y la naturaleza jurídica (oficial o no oficial, COLE_NATURALEZA). Estas variables permiten analizar las características sociodemográficas, educativas y del entorno que pueden influir en el desempeño académico.

```python
filtro_periodo = [20221, 20224]
nuevo_df2 = nuevo_df[nuevo_df['PERIODO'].isin(filtro_periodo)]
nuevo_df2.shape
```

```python
# Análisis de la variable objetivo
print(nuevo_df2['PUNT_GLOBAL'].describe())
nuevo_df2['PUNT_GLOBAL'].plot.hist()
```

```python
# dumizar
cols_dummy = [
    'COLE_AREA_UBICACION',
    'COLE_CALENDARIO',
    'COLE_CARACTER',
    'COLE_JORNADA',
    'COLE_GENERO',
    'COLE_NATURALEZA',
    'ESTU_GENERO',
    'FAMI_CUARTOSHOGAR',
    'FAMI_EDUCACIONMADRE',
    'FAMI_EDUCACIONPADRE',
    'FAMI_ESTRATOVIVIENDA',
    'FAMI_PERSONASHOGAR',
    'FAMI_TIENEAUTOMOVIL',
    'FAMI_TIENECOMPUTADOR',
    'FAMI_TIENEINTERNET',
    'FAMI_TIENELAVADORA',
]
nuevo_df3 = pd.get_dummies(nuevo_df2, columns=cols_dummy)
nuevo_df3.drop(columns=['ESTU_FECHANACIMIENTO', 'PERIODO'], inplace=True)
nuevo_df3
```

```python
# Agrupar por la nombre de la sede
nuevo_df4 = nuevo_df3.groupby(['COLE_NOMBRE_SEDE']).mean()
nuevo_df4
```

```python
# Define el nombre de la sede que te interesa
nombre_sede = "COL DIST TOMAS CARRASQUILLA"  # Reemplaza con el nombre de la sede que buscas

# Obtén el PUNT_GLOBAL para esa sede
puntaje = nuevo_df4.loc[nombre_sede, 'PUNT_GLOBAL']

# Imprime el puntaje
print(f"El PUNT_GLOBAL para {nombre_sede} es: {puntaje}") 
```

```python
cole_ref = 'COL DIST TOMAS CARRASQUILLA'
nuevo_df4[nuevo_df4.index == cole_ref].T
```

```python
# Crear modelo pca y graficar en el plano las dos primeras componentes
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = nuevo_df4.drop(columns=['PUNT_GLOBAL'])
y = nuevo_df4['PUNT_GLOBAL']

modelo_pca = PCA(n_components=2)
modelo_pca.fit(X)
componentes_pca = modelo_pca.transform(X)

# Obtener la varianza explicada por cada componente
varianza_explicada = modelo_pca.explained_variance_ratio_

# Crear el gráfico de dispersión
sns.scatterplot(x=componentes_pca[:, 0], y=componentes_pca[:, 1], 
                alpha=0.5, color='#12A2A8')

# Añadir etiquetas a los ejes con la varianza explicada
plt.xlabel(f'Componente Principal 1 ({varianza_explicada[0]*100:.1f}%)')
plt.ylabel(f'Componente Principal 2 ({varianza_explicada[1]*100:.1f}%)')


# sns.scatterplot(x=componentes_pca[:, 0], y=componentes_pca[:, 1], 
#                 alpha=0.5,  
#                 color='#12A2A8' 
#                )
```

```python
import altair as alt

# Obtener los loadings (pesos de las variables en cada componente)
loadings = pd.DataFrame(modelo_pca.components_.T, columns=['PC1', 'PC2'], index=X.columns)


# Calcular la varianza de cada variable en cada componente
varianza_variables = loadings ** 2 * varianza_explicada

# Mostrar la varianza de cada variable en cada componente
print("Varianza de cada variable en cada componente:")
print(varianza_variables)

# --- Visualización con Altair ---

# Preparar los datos para Altair
#varianza_variables_melted = varianza_variables.reset_index().melt(id_vars='index')
#varianza_variables_melted = varianza_variables_melted.rename(columns={'index': 'Variable', 'variable': 'Componente', 'value': 'Varianza'})

# # Crear el gráfico de barras con Altair
# chart = alt.Chart(varianza_variables_melted).mark_bar().encode(
#     x=alt.X('Variable', axis=alt.Axis(title='Variables')),
#     y=alt.Y('Varianza', axis=alt.Axis(title='Varianza')),
#     color='Componente',
#     tooltip=['Variable', 'Componente', 'Varianza']
# ).properties(
#     title="Varianza de cada variable en los componentes principales"
# ).interactive()

# # Mostrar el gráfico en el navegador
# chart.show()
```

```python
# --- Mostrar las variables que más aportan ---
numero_de_variables = 5

# Ordenar los loadings por valor absoluto para PC1 y mostrar las primeras variables
print("Variables que más aportan a PC1:")
print(loadings.iloc[loadings['PC1'].abs().argsort()[::-1]][:numero_de_variables])

# Ordenar los loadings por valor absoluto para PC2 y mostrar las primeras variables
print("\nVariables que más aportan a PC2:")
print(loadings.iloc[loadings['PC2'].abs().argsort()[::-1]][:numero_de_variables])
```

```python
# Gráfico para todos los puntos
sns.scatterplot(x=componentes_pca[:, 0], y=componentes_pca[:, 1], 
                alpha=0.3,  # Transparencia 
                color='#12A2A8'  # Color de los puntos generales
               )

# Resaltar el colegio 'cole_ref' 
sns.scatterplot(x=componentes_pca[nuevo_df4.index == cole_ref, 0], 
                y=componentes_pca[nuevo_df4.index == cole_ref, 1], 
                color='#FF7F0E', # Color para el colegio específico
                label = "Tomás Carrasquilla"
               )

# Añadir etiquetas a los ejes con la varianza explicada
plt.xlabel(f'Componente Principal 1 ({varianza_explicada[0]*100:.1f}%)')
plt.ylabel(f'Componente Principal 2 ({varianza_explicada[1]*100:.1f}%)')

plt.title('PCA para los colegios de Bogotá')
plt.legend(loc='lower right')
# Guardar el gráfico en formato PNG
#plt.savefig('img/PCA_tc.svg')
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import silhouette_samples

siluetas=[]

for k in range(1,50):
    
    modelo_nn = NearestNeighbors(n_neighbors=k)
    modelo_nn.fit(componentes_pca)
    distancias, indices = modelo_nn.kneighbors(componentes_pca)

    col_cercanos= indices[nuevo_df4.index == cole_ref][0]
    y_pred = np.zeros(len(X),dtype=int)
    y_pred[col_cercanos]=1

    silhouette_coefficients = silhouette_samples(X, y_pred)
    silueta_modelo_k = silhouette_coefficients[col_cercanos[0]]
    siluetas.append(silueta_modelo_k)

# Crear un DataFrame con los datos
df_siluetas = pd.DataFrame({'K vecinos más cercanos': range(1, 50), 'Puntaje Silueta': siluetas})

# Graficar con seaborn
sns.lineplot(x='K vecinos más cercanos', 
             y='Puntaje Silueta', 
             data=df_siluetas,
            color ='#12A2A8')
plt.title('Puntaje silueta entre 0 y 150 Colegios (k)')
#plt.savefig('img/puntaje_silueta_150.svg')
plt.show()
#plt.plot(siluetas)
#siluetas
```
