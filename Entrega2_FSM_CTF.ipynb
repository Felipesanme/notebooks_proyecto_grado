## 3. Antecedentes
    
    Para este trabajo se tuvieron en cuenta estudios anteriores realcionados con el estudio de Pruebas Saber 11 y Saber Pro que utilizando tecnicas de machine Learning como el K-means, arboles de desicion, redes neuronales, han logrado evidenciar la influencia de la estadistica aplicada en el campo educativo. 
    
    Carrascal y Giraldo (2019) realizaron un estudio titulado Minería de Datos Educativos: Análisis del Desempeño de Estudiantes de Ingeniería en las Pruebas Saber-Pro. El objetivo de este trabajo fue contribuir al análisis de las pruebas Saber-Pro en Colombia, examinando variables académicas, económicas, sociodemográficas y los resultados de todos los módulos de dichas pruebas correspondientes al año 2016, específicamente de estudiantes de ingeniería en Antioquia. Para ello, emplearon tres modelos analíticos de clustering, identificaron los factores más influyentes en el desempeño de cada módulo de las pruebas de manera individual, y predijeron los resultados en función de las variables seleccionadas.

    Entre los hallazgos más destacados, se identificó que el género fue un factor determinante en los módulos de razonamiento cuantitativo y comunicación escrita, aunque tuvo menor incidencia en el desempeño de los demás módulos. Además, el número de libros en casa resultó ser un factor clave para la comunicación escrita, mientras que la condición económica influyó significativamente en el módulo de inglés. Por otro lado, se encontró que ser cabeza de familia representaba una barrera importante para obtener buenos resultados en las pruebas. Finalmente, el estudio también evidenció que los estudiantes que cursan sus estudios de manera presencial tienden a obtener mejores resultados, un aspecto que coincide con los hallazgos de investigaciones previas.

    Timarán, Caicedo y Hidalgo (2019) llevaron a cabo un estudio titulado Árboles de Decisión para Predecir Factores Asociados al Desempeño Académico de Estudiantes de Bachillerato en las Pruebas Saber 11°. El objetivo fue identificar los factores que influyen en el desempeño académico de los estudiantes colombianos de grado undécimo que presentaron las pruebas Saber 11° en los años 2015 y 2016.

    Para descubrir patrones relacionados con el desempeño académico, los investigadores desarrollaron un modelo de clasificación basado en árboles de decisión, empleando el algoritmo J48 de la herramienta WEKA y siguiendo la metodología CRISP-DM. El modelo, construido con la base de datos T1061680A17, logró clasificar correctamente 711,116 instancias, alcanzando una precisión del 67%, mientras que 350,564 instancias fueron clasificadas incorrectamente, representando un 33% de error.

    Los autores concluyeron que los árboles de decisión son herramientas efectivas para generar modelos consistentes con la realidad observada, respaldados teóricamente y basados únicamente en los datos almacenados en las bases de datos del ICFES. Este enfoque proporciona valiosa información para analizar y predecir el desempeño académico en las pruebas Saber 11°.

    Rojas (2021) llevó a cabo un estudio titulado Comparación de Herramientas Libres de Minería de Datos Educativa que Aplican Técnicas Descriptivas. Caso de Estudio: Caracterización Socioeconómica de los Estudiantes en las Pruebas Saber 11 de 2018 y 2019. El objetivo fue comparar herramientas de minería de datos educativa de código abierto que aplican técnicas descriptivas, con el propósito de caracterizar la situación socioeconómica de los estudiantes que presentaron las pruebas Saber 11 en los años mencionados.

    Para ello, implementó una metodología comparativa de las herramientas KNIME, Orange y WEKA, siguiendo el proceso de minería de datos educativa guiado por la metodología CRISP-DM. Entre los hallazgos, se identificaron como factores clave para un mejor desempeño académico el acceso a bienes y servicios, particularmente un computador y conexión a internet, así como el nivel educativo de los padres. Los estudiantes cuyos padres alcanzaron niveles educativos superiores al bachillerato obtuvieron mejores puntajes. Por otro lado, aquellos con bajo rendimiento presentaron características como mayor edad, convivir con más de siete personas en el hogar, poseer motocicleta, trabajar más de 10 horas a la semana y recibir remuneración en efectivo.

    La implementación de los algoritmos K-Means y Jerárquico en las tres herramientas mostró resultados consistentes, agrupando los clústeres según los puntajes obtenidos en las diferentes áreas de las pruebas (bajos, medios y altos) y las variables relacionadas con dichos puntajes. Este enfoque destacó la capacidad de las herramientas para analizar y categorizar factores socioeconómicos relevantes en el desempeño académico.
